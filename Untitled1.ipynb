```python
import pandas as pd
df= pd.read_csv("/content/marketing_campaign.csv" ,sep="\t")
print(df.shape)
df.head()
```

```python
# Check missing values
print(df.isnull().sum())

df["Income"] = df["Income"].fillna(df["Income"].median())

```

```python
# removing duplicate values
df=df.drop_duplicates()

df["Education"] = df["Education"].str.strip().str.lower()
df["Marital_Status"] = df["Marital_Status"].str.strip().str.lower()
edu_map = {
    "2n cycle": "masters",
    "basic": "high school",
    "graduation": "bachelors",
    "master": "masters",
    "phd": "phd"
}

marital_map = {
    "together": "partnered",
    "married": "married",
    "single": "single",
    "divorced": "divorced",
    "widow": "widowed",
    "alone": "single",   # optional choice
    "absent": "single",  # optional
    "yolo": "single"     # optional
}

df["Education"] = df["Education"].replace(edu_map)
df["Marital_Status"] = df["Marital_Status"].replace(marital_map)

```

```python
edu_map = {
    "2n cycle": "masters",
    "basic": "high school",
    "graduation": "bachelors",
    "master": "masters",
    "phd": "phd"
}

marital_map = {
    "together": "partnered",
    "married": "married",
    "single": "single",
    "divorced": "divorced",
    "widow": "widowed",
    "alone": "single",   # optional choice
    "absent": "single",  # optional
    "yolo": "single"     # optional
}

df["Education"] = df["Education"].replace(edu_map)
df["Marital_Status"] = df["Marital_Status"].replace(marital_map)
# Convert to datetime
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], dayfirst=True, errors="coerce")
```

```python
df.columns = (
    df.columns.str.strip()
              .str.lower()
              .str.replace(" ", "_")
)

print(df.columns.tolist())

df["year_birth"] = df["year_birth"].astype(int)

df["age"] = pd.Timestamp.today().year - df["year_birth"]

df["income"] = df["income"].astype(float)


campaign_flags = [c for c in df.columns if "acceptedcmp" in c.lower()] + ["response", "complain"]
df[campaign_flags] = df[campaign_flags].astype(int)
```
